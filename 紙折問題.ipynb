```python
# 【問題1】べき乗の算術演算子を使用した実装

# 紙を43回折った時の厚さを計算するコード

THICKNESS = 0.00008

folded_thickness = THICKNESS * (2 ** 43)

print("厚さ： {}メートル".format(folded_thickness))

# 【問題2】単位の変換

# メートルを万キロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 10000000))

# 【問題3】for文を使用して作成

THICKNESS = 0.00008
result = 1
for num in range(1, 44):
 result *= 2

answer = result * THICKNESS
print("厚さ： {:.2f}万キロメートル".format(answer / 10000000))

# 【問題4】計算時間の比較

# 【問題1】の速度計算
import time
start = time.time()
olded_thickness = THICKNESS * (2 ** 43)
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))

# 【問題3】の速度計算
import time
start = time.time()
result = 1
for num in range(1, 44):
 result *= 2

answer = result * THICKNESS
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))

# 【問題5】リストへの保存

paper = []
THICKNESS = 0.00008
paper.append(THICKNESS)

result = 1

for num in range(1, 44):
 result *= 2
 paper.append(result * THICKNESS)

print(paper)

# 【問題6】折れ線グラフの表示

import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(paper)
plt.show()

# 40回以上折った時点から急激に暑さが大きくなる

#【問題7】グラフのカスタマイズ

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(paper, '--' ,lw = 10.0, color = 'green')
plt.show()
```
